```python
#Filtering Dictionaries in a List
all_cafes = [
    {"name": "Bondi Cafe", "open": True},
    {"name": "Manly Cafe", "open": False},
    {"name": "Newtown Cafe", "open": True}
]

cafes = [cafe for cafe in all_cafes if cafe.get("open")]
print(cafes)
```

```python
#Find Closed Cafes
all_cafes = [
    {"name": "Bondi Cafe", "open": True},
    {"name": "Manly Cafe", "open": False},
    {"name": "Newtown Cafe", "open": True}
]

closed_cafes = [cafe for cafe in all_cafes if not cafe.get("open")]
print(closed_cafes)
```

```python
#Student Enrolment System
#Define student data using tuples (ID, Name)
students = [(101, "Toby"), (102, "Farshid"), (103, "Nolan")]
#Define available workshops
workshops = ["Python", "AI", "Data Ethics", "Cybersecurity"]
#Dictionary to map student ID to their selected workshops
enrolments = {}
#Enrol students using input and control flow
for student_id, name in students:
  print(f"\nEnrolling {name} (ID: {student_id})")
  print("Available workshops:", workshops)
  selected = input(f"{name}, choose 2 workshops (comma-seperated): ").split(",")
  #Remove duplicates using set and keep only valid workshops
  selected_set = set([w.strip() for w in selected if w.strip() in workshops])
  #Ensure they choose 2 different workshops
  if len(selected_set) != 2:
    print("you must choose 2 **different** valid workshops.")
    continue
  #Save the selection
  enrolments[student_id] = selected_set
#Add a new student
students.append((104, "Liam"))
#Add a new workshop
workshops.append("Robotics")
#Re-run enrolment for the new student
print(f"\nEnrolling Liam (ID:104)")
print("Available workshops:", workshops)
selected = input("Liam, choose 2 workshops (comma-seperated): ").split(",")
selected_set = set([w.strip() for w in selected if w.strip() in workshops])
if len(selected_set) == 2:
  enrolments[104] = selected_set
else:
  print("You must choose 2 **different** valid workshops.")
#Display all enrolments
print("\n Final Enrolment Summary:")
for student_id, workshops_set in enrolments.items():
  name = [s[1] for s in students if s[0] == student_id][0]
  print(f"{name} ({student_id}): {', '.join(workshops_set)}")
#Count workshop popularity
workshop_counts = {}
for enrolled_set in enrolments.values():
  for w in enrolled_set:
    workshop_counts[w] = workshop_counts.get(w, 0) + 1
print("\nWorkshop Popularity:")
for w, count in workshop_counts.items():
  print(f"{w}: {count} student(s)")
```
